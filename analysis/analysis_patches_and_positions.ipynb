# Analyse how shuffling influences the score

```python
"""
Function to shuffle an list of galaxies
Arguments: 
- galaxies - list of galaxies

Output: 
- shuffled list of galaxies 
"""

def shuffle_galaxies(galaxies):
    new_location = np.arange(len(galaxies))
    np.random.shuffle(new_location)

    shuffled_galaxies = [None] * len(galaxies)

    for i in range(len(galaxies)):
        shuffled_galaxies[i] = galaxies[new_location[i]]
    
    return shuffled_galaxies
```

```python
"""
Function to generate a cosmologic image using patches and a fixed position matrix
Arguments:
- position - position matrix generated
- galaxies - list of patches with generated galaxies
- n_shuffles - number of images to generate with these settings by shuffling where the patches are put on the position matrix
"""
def generate_on_fixed_position(position_matrix, galaxies, n_shuffles=1):
    assert n_shuffles > 0
    
    imgs = []
    position = list(zip(*np.where(position_matrix==1)))
        
    for i_shuffles in range(n_shuffles):
        galaxies = shuffle_galaxies(galaxies)
        cur_galaxy = 0
        new_img = np.zeros((1024, 1024))

        for location in position:
            top = int(location[0]*patch_size[0])
            bottom = int((location[0]+1)*patch_size[0])
            left = int(location[1]*patch_size[1])
            right = int((location[1]+1)*patch_size[1])
            new_img[top:bottom, left:right] = galaxies[cur_galaxy]
            cur_galaxy += 1

        imgs.append(new_img[:image_height, :image_width])
    return imgs
```

```python
def check_shuffle_patches(n_positions, n_shuffles):
    cols = ["Average Score", "Max Score", "Min Score", "Max-Min", "Number of Galaxies"]
    df = pd.DataFrame(columns=cols)
    
    for _ in range(n_positions):
        # generate one position matrix
        positions, n_galaxies = position_gan.generate(1)

        # generate required number of galaxies
        galaxies = galaxy_gan.generate(n_galaxies)

        # generate images using this position matrix by shuffling where the galaxies are put
        generated_images = generate_on_fixed_position(positions[0], galaxies, n_shuffles)

        # evaluate the generated images
        scores = regressor.score(generated_images)

        # create entry for recording data that could be useful
        # - min score, max score, average score, number of galaxies in image
        min_score = np.min(scores)
        max_score = np.max(scores)
        avg_score = np.sum(scores) / len(scores)

        # create row for dataset and add it
        new_row = pd.DataFrame([[avg_score, max_score, min_score, max_score - min_score, n_galaxies]], columns=cols)
        df = df.append(new_row)
    
    return df
```

```python
n_positions = 1000
n_shuffles = 50

df = check_shuffle_patches(n_positions = n_positions, n_shuffles = n_shuffles)
```

# Analyse how the patches choice influences the score 
- for a fixed position matrix, vary the patches used

```python
"""
Function to generate images on a given fixed position matrix 
the patches used to generate image are completely changed each time
"""
def generate_on_fixed_position_change_patches(position_matrix, n_galaxies, n_sets_patches):
    # create necessary number of patches
    galaxies = galaxy_gan.generate(n_galaxies * n_sets_patches)
    
    cur_galaxy = 0
    imgs = []
    
    position = list(zip(*np.where(position_matrix==1)))
    
    for n in range(n_sets_patches):
        new_img = np.zeros((1024, 1024))

        for location in position:
            top = int(location[0]*patch_size[0])
            bottom = int((location[0]+1)*patch_size[0])
            left = int(location[1]*patch_size[1])
            right = int((location[1]+1)*patch_size[1])
            new_img[top:bottom, left:right] = galaxies[cur_galaxy]
            cur_galaxy += 1

        imgs.append(new_img[:image_height, :image_width])
        
    return imgs
```

```python
def check_change_patches(n_positions, n_sets_patches):
    cols = ["Average Score", "Max Score", "Min Score", "Max-Min", "Number of Galaxies"]
    df = pd.DataFrame(columns=cols)

    
    for _ in range(n_positions):
        # generate one position matrix
        positions, n_galaxies = position_gan.generate(1)
        
        # vary patches used for this position configuration 
        generated_images = generate_on_fixed_position_change_patches(positions[0], n_galaxies, n_sets_patches)

        # evaluate the generated images
        scores = regressor.score(generated_images)

        # create entry for recording data that could be useful
        # - min score, max score, average score, number of galaxies in image
        min_score = np.min(scores)
        max_score = np.max(scores)
        avg_score = np.sum(scores) / len(scores)

        # create row for dataset and add it
        new_row = pd.DataFrame([[avg_score, max_score, min_score, max_score - min_score, n_galaxies]], columns=cols)
        df = df.append(new_row)
    
    return df
```

```python
n_positions = 1000
n_sets_patches = 50 

df_2 = check_change_patches(n_positions=n_positions, n_sets_patches=n_sets_patches)
```
